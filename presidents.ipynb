```python
import re
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import plotly.express as px

def get_soup(input_url,parser,find):
    url = input_url
    r = requests.get(url)
    if find == "content" :
        soup = BeautifulSoup(r.content, parser)
    else :
        soup = BeautifulSoup(r.text, parser)
    return soup
```

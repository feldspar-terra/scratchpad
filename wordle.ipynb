```python
import pandas as pd
import requests
```

```python
url = 'https://raw.githubusercontent.com/dwyl/english-words/master/words_dictionary.json'
response = requests.get(url)
context = pd.json_normalize(response.json()).T.reset_index().rename(columns={"index":"word",0:"drop"}).drop(columns=['drop'])
context['length'] = context.word.str.len()
five_letter_words = context[context['length']==5]
five_letter_words
```

```python
letter_ordinal = five_letter_words['word'].str.split('', expand=True)[[1,2,3,4,5]]
letter_ordinal
```

```python
def add_ordinal_col(df,ordinal):
    ordinal_df = df.groupby([ordinal])[ordinal].count().to_frame().rename(columns={ordinal:"count"})
    ordinal_df['ordinal'] = ordinal
    return ordinal_df

    # 1 letter_ordinal
```

```python
pd.concat([
    add_ordinal_col(letter_ordinal,1),
    add_ordinal_col(letter_ordinal,2),
    add_ordinal_col(letter_ordinal,3),
    add_ordinal_col(letter_ordinal,4),
    add_ordinal_col(letter_ordinal,5)
]).reset_index().pivot_table(columns=['ordinal'],values=['count'],index=['index'])
```
